**C**
```c
#include <stdio.h>
#include <string.h>

int main() {
    char text[] = "Python je vysokoúrovňový programovací jazyk";
    int count = 0;
    char *token = strtok(text, " ");

    while (token != NULL) {
        count++;
        token = strtok(NULL, " ");
    }

    printf("Pocet slov: %d\n", count);
    return 0;
}
```

### Interpretovaný Jazyk

Program může být spuštěn dvěma hlavními způsoby:

1. Kompilací
2. Interpretací

Rozdíl spočívá v tom, kdy a jak se zdrojový kód překládá do strojového kódu.


**Kompilovaný jazyk**

Zdrojový kód → kompilátor → strojový kód → spuštění

Program se přeloží celý před spuštěním.

Charakteristiky:
- rychlý běh
- chyby zachyceny při kompilaci
- vznikne samostatný spustitelný soubor


**Interpretovaný jazyk**

Zdrojový kód → interpret → okamžité vykonání

Kód se překládá a vykonává postupně za běhu.

Charakteristiky:
- není nutná ruční kompilace
- rychlé testování
- chyby se projeví při běhu programu



**Interaktivní Prostředí (REPL):**  
  Python nabízí interaktivní interpret (Read-Eval-Print Loop), kde můžete psát jednotlivé příkazy a okamžitě vidět jejich výsledek. To je skvělé pro:
  - Experimentování s novými nápady.
  - Rychlé testování funkcí a algoritmů.
  - Učení se základům jazyka v reálném čase.

### Podpora Různých Programovacích Paradigmat

- **Procedurální Programování:**  
  Umožňuje psát sekvenční bloky kódu, kde jsou příkazy vykonávány jeden po druhém. Je to jednoduchý způsob, jak řešit menší úlohy.

- **Objektově Orientované Programování (OOP):**  
  Python podporuje OOP, což znamená, že můžete definovat třídy a objekty. To pomáhá:
  - Modelovat složité systémy.
  - Znovu použít a rozšiřovat kód pomocí dědičnosti a polymorfismu.

- **Funkcionální Programování:**  
  Python obsahuje prvky funkcionálního programování, jako jsou lambda funkce, map, filter a list comprehension.

### Bohatá Standardní Knihovna

- **Vestavěné Moduly a Balíčky:**  
  Python přichází s rozsáhlou standardní knihovnou, která obsahuje moduly pro různé úlohy:
  - **Práce se Souborovým Systémem:** čtení a zápis souborů.
  - **Komunikace po Síti:** HTTP klienti, socket programování.
  - **Datová Analýza a Zpracování:** práce s daty, statistické funkce a mnoho dalšího.

- **Rozšiřitelnost:**  
  Kromě standardní knihovny existuje obrovská komunita, která vytváří další knihovny a frameworky. Díky tomu najdete řešení pro téměř každou oblast, ať už jde o webový vývoj, datovou vědu, strojové učení nebo automatizaci.

### Odsazení v Pythonu

V Pythonu se bloky kódu (např. obsah podmínky, tělo funkce, cykly) určují pomocí odsazení, nikoli pomocí závorek jako v jiných jazycích (např. `{}` v C nebo Java). Správné odsazení je tedy klíčové pro správnou strukturu a sémantiku vašeho kódu.

#### Základní pravidla

- **Konzistence:**  
  V rámci jednoho bloku musí být použito stejné odsazení (obvykle se doporučuje 4 mezery). Smíchat mezery a tabulátory se nedoporučuje, protože to může vést k nejednoznačnosti a chybám.

- **Určení bloku:**  
  Blok kódu začíná řádkem, který je odsazený více než předcházející řádek. Konec bloku je určen tím, že se řádek vrátí na nižší úroveň odsazení.

## Řídicí struktury a cykly

Podmíněné příkazy (if, elif, else) a cykly (for, while) jsou základními stavebními kameny pro řízení toku programu.

````python
# Podmíněný příkaz
x = 9
if x > 10:    
    print('x je větší než 10')        
elif x == 10:
    print('x je rovno 10')
else:
    print('x je menší než 10')

````

### Cyklus for
Iteruje přes položky jakékoliv sekvence (např. seznam, n-tice, slovník).

````python
kosik = ["jablko", "banán", "mango"]
for ovoce in kosik:
    print(ovoce)
````

Pro specifický počet opakování se používá funkce range()

````python
for i in range(10):
    print(i)
````

### Cyklus while

Opakuje blok kódu tak dlouho, dokud je splněna určitá podmínka.

````python
i = 0
while i < 5:
    print(i)
    i += 1
````

### Přerušení cyklu: break a continue

Příkaz **break** okamžitě ukončí celý cyklus.

````python
for i in range(10):
    if i == 5:
        break # Ukonci cyklus, kdyz i dosahne 5
    print(i)
````

Příkaz **continue** ukončí aktuální iteraci cyklu a pokračuje další iterací.

````python
for i in range(10):
    if i == 5:
        continue # Preskoci 5 a pokracuje dale
        print('Tento string se nevypise.')
    print(i)
````

## Funkce

### Úvod
- Funkce v Pythonu jsou bloky kódu, které jsou navrženy k vykonání konkrétního úkolu a mohou být opakovaně volány z různých částí programu. 

- Funkce se v Pythonu deklaruje pomocí klíčového slova **def**, následovaného názvem funkce a kulatými závorkami, které mohou obsahovat parametry. 

- Tělo funkce je **odsazeno** a začíná hned pod řádkem deklarace.

````python
def pozdrav(jmeno):         #Definice funkce
    print("Ahoj, " + jmeno + "!")


pozdrav("Alice")    #Volani funkce
pozdrav("Jakube")
````

### Parametry a argumenty
- Parametry jsou proměnné uvedené v definici funkce.

- Argumenty jsou hodnoty předané funkci při jejím volání.

````python
def pozdrav(jmeno, pozdrav="Cau"): #Vychozi hodnota parametru
    print(pozdrav + ", " + jmeno + "!")
    
    
pozdrav('Adame')
````

- Funkce mohou vrátit hodnoty použitím příkazu **return**. 

- Pokud funkce nemá return příkaz, implicitně vrací **None**.

````python


def soucet(a, b):
    res = a + b
    return res   #Vraci vysledek souctu promennych a,b

x = 5
soucet(2,x)
````
